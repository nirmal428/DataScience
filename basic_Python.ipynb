```python
#Break and Continue

stu=["ram","shyam","jaya","amit","jyoti"]
for student in stu:
  if student=="amit":
    break;
  print(student)

stu=["ram","shyam","jaya","amit","jyoti"]
for student in stu:
  if student=="amit":
    continue;
  print(student)
```
